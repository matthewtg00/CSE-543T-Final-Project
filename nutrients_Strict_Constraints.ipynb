```python
def preference_factor(row):
    if row['Category'] == 'Dairy products':
        return 0.4
    if row['Category'] == 'Fats, Oils, Shortenings':
        return 0.3
    if row['Category'] == 'Meat, Poultry':
        return 0.8
    if row['Category'] == 'Fish, Seafood':
        return 0.6
    if row['Category'] == 'Vegetables A-E':
        return 0.4
    if row['Category'] == 'Vegetables F-P':
        return 0.4
    if row['Category'] == 'Vegetables R-Z':
        return 0.4
    if row['Category'] == 'Fruits A-F':
        return 0.6
    if row['Category'] == 'Fruits G-P':
        return 0.6
    if row['Category'] == 'Fruits R-Z':
        return 0.6
    if row['Category'] == 'Breads, cereals, fastfood,grains':
        return 0.5
    if row['Category'] == 'Soups':
        return 0.2
    if row['Category'] == 'Desserts, sweets':
        return 0.8
    if row['Category'] == 'Jams, Jellies':
        return 0.5
    if row['Category'] == 'Drinks,Alcohol, Beverages':
        return 0.6
```

```python
df['preference_factor'] = df.apply(preference_factor, axis=1)
df
```

```python
import numpy as np
import scipy.optimize as opt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def calculate_required_calories(weight=None, height=None, age=None, gender=None, activity_factor=None):
    # Mifflin -St Jeor Equations(MSJE)
    required_cal=2000 #default required 2000 calorie
    if weight is not None and height is not None and age is not None and gender is not None and activity_factor is not None:
        if gender=="male":
            required_cal=(10*weight+6.25*height-5*age+5)*activity_factor
        elif gender=="female":
            required_cal=(10*weight+6.25*height-5*age-161)*activity_factor
    return required_cal
```

```python
from functools import partial
def diet_optimization(x, required_calories):
    # carb 45~60% 1carb=4cal
    diff_carb=abs(0.525*required_calories-4*np.dot(df['Carbs'].T,x))

    # protein 25~35% 1protein=4cal
    diff_protein=abs(0.3*required_calories-4*np.dot(df['Protein'].T,x))

    # fat 20~35% 1fat=9cal
    diff_fat=abs(0.225*required_calories-9*np.dot(df['Fat'].T,x))
    # calorie = np.sum(np.multiply(df['Calories'], x**2))
    # fat = np.sum(np.multiply(df['Fat'], x**2))*10
    # carbs = np.sum(np.multiply(df['Carbs'], x**2))*10
    return diff_carb+diff_protein+diff_fat


required_calories=calculate_required_calories(83,180,23,"male",1.3)

partial_diet_optimization = partial(diet_optimization, required_calories=required_calories)
```

```python
def lower_calorie_constraint(x, required_calories):
    # gap between cal and required_cal should no more than 500
    calorie = np.dot(df['Calories'].T, x)
    return (calorie-required_calories+500)

def upper_calorie_constraint(x, required_calories):
    # gap between cal and required_cal should no more than 300
    calorie = np.dot(df['Calories'].T, x)
    return -(calorie-required_calories-300)

# def protein_constraint(x):
#     protein = np.sum(np.multiply(df['Protein'], x))
#     return (protein-100)

# def fiber_constraint(x):
#     fiber = np.sum(np.multiply(df['Fiber'], x))
#     return (fiber-15)

def taste_constraint(x):
    taste = np.sum(x**df['taste_factor'])
    return (taste-10)

def perference_constraint(x):
    preference = np.sum(np.multiply(df['preference_factor'], x))
    return (preference-10)

def volume_constraint(x):
    volume = np.sum(np.multiply(df['Grams'], x))
    return -(volume-2000)

# At most choose 10 items
def items_constraint_upper(x):
    return -(2*np.sum(sigmoid(10*x)-0.5)-9.999999)
    

partial_lower_calorie_constraint = partial(lower_calorie_constraint, required_calories=required_calories)
partial_upper_calorie_constraint = partial(upper_calorie_constraint, required_calories=required_calories)

constraints = [{'type': 'ineq', 'fun': partial_lower_calorie_constraint},
               {'type': 'ineq', 'fun': partial_upper_calorie_constraint},
               {'type': 'ineq', 'fun': volume_constraint},
            #    {'type': 'ineq', 'fun': protein_constraint},
            #    {'type': 'ineq', 'fun': fiber_constraint},
               {'type': 'ineq', 'fun': taste_constraint},
               {'type': 'ineq', 'fun': perference_constraint},
               {'type': 'ineq', 'fun': items_constraint_upper}]

bounds = [(0,2)] * df.shape[0]
```

### SLSQP (Sequential Least Square Programming)

```python
x0 = np.zeros(df.shape[0])
options = {"disp": True, "maxiter": 5000}
result = opt.minimize(partial_diet_optimization, x0, method='SLSQP', bounds=bounds, 
                      constraints=constraints,options=options)
print(result.x)
```

```python
calorie = np.sum(np.multiply(df['Calories'], result.x))
fat = np.sum(np.multiply(df['Fat'], result.x))
carbs = np.sum(np.multiply(df['Carbs'], result.x))
protein = np.sum(np.multiply(df['Protein'], result.x))
taste = np.sum(result.x**df['taste_factor'])
preference = np.sum(np.multiply(df['preference_factor'], result.x))
grams = (np.multiply(df['Grams'], result.x))

print(required_calories)
print(calorie)
print(protein)
print((result.x>0.01).sum())
print(taste)
print(preference)
df['diet'] = result.x
```

```python
df['Number of Measure'] = result.x
filtered_df = df[df['Number of Measure'] >= 0.01][['Food', 'Measure', 'Number of Measure']]
pd.set_option('display.float_format', '{:.2f}'.format)
filtered_df
```

### Trust - Contr Solver (Trust Region Constraint)

```python
result_trust_constr = opt.minimize(partial_diet_optimization, x0, method='trust-constr', bounds=bounds, 
                                   constraints=constraints, options=options)
```

```python
# Display the results
print(result_trust_constr.x)

print("\nOptimization success:", result_trust_constr.success)
print("Objective function value at the solution (minimized difference):", result_trust_constr.fun)
print("Number of iterations performed:", result_trust_constr.nit)
print("Message:", result_trust_constr.message)
```

```python
calorie = np.sum(np.multiply(df['Calories'], result_trust_constr.x))
fat = np.sum(np.multiply(df['Fat'], result_trust_constr.x))
carbs = np.sum(np.multiply(df['Carbs'], result_trust_constr.x))
protein = np.sum(np.multiply(df['Protein'], result_trust_constr.x))
taste = np.sum(result.x**df['taste_factor'])
preference = np.sum(np.multiply(df['preference_factor'], result_trust_constr.x))
grams = (np.multiply(df['Grams'], result_trust_constr.x))

print(required_calories)
print(calorie)
print(protein)
print((result_trust_constr.x>0.01).sum())
print(taste)
print(preference)
```

```python
df['Number of Measure'] = result_trust_constr.x
filtered_df = df[df['Number of Measure'] >= 0.01][['Food', 'Measure', 'Number of Measure']]
pd.set_option('display.float_format', '{:.2f}'.format)
filtered_df
```

### COBYLA (Constrained Optimization by Linear Approximation)

```python
from scipy.optimize import minimize
options = {"maxiter": 5000}

cobyla_constraints = constraints + [{'type': 'ineq', 'fun': lambda x, lb=lb, ub=ub: x - lb, 'args': (lb,)} for lb, ub in bounds] + \
                                  [{'type': 'ineq', 'fun': lambda x, lb=lb, ub=ub: ub - x, 'args': (ub,)} for lb, ub in bounds]

# Perform the optimization using the 'COBYLA' method
result_cobyla = minimize(partial_diet_optimization, x0, method='COBYLA',
                         constraints=cobyla_constraints, options=options)

print(result_cobyla.x)

print("\nOptimization success:", result_cobyla.success)
print("Objective function value at the solution (minimized difference):", result_cobyla.fun)
# COBYLA does not provide a standardized 'nit' attribute for the number of iterations
print("Message:", result_cobyla.message)
```

```python
result_tnc = opt.minimize(partial_diet_optimization, x0, method='TNC', bounds=bounds, 
                          constraints=constraints, options=options)

print("\nOptimization success:", result_tnc.success)
print("Objective function value at the solution (minimized difference):", result_tnc.fun)
print("Number of iterations performed:", result_tnc.nit)
print("Message:", result_tnc.message)
```
